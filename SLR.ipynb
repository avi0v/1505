```python
#IMPORTING LIBS
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import statsmodels.formula.api as smf 
```

```python
#LOADING DATA
data=pd.read_csv("NewspaperData.csv")
data.head()
```

```python
#scatter plot to check pattern
sns.scatterplot(x='daily',y='sunday',data=data)
```

```python
#finding best fit line or MODEL
#y=mx+c
model=smf.ols("daily~sunday",data=data).fit()
```

```python
#plotting 
sns.regplot(x='daily',y='sunday',data=data)
```

```python
#m,c calculate
model.params
```

```python
#fitter value or calculted value or predicted value
model.fittedvalues
```

```python
#error = actual - fitted or actual - calculted from model
model.resid
```

```python
# rsquare calculate
model.rsquared
```

```python
#conclusion r2 - 91% so good fit
#r2-variation of model/variation of actual from mean
```

```python
#pvalue calulate 
model.pvalues
```
